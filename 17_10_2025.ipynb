 # LIST METHODS

```python
"""como hacer una lista xd"""

lista = ["luque","hugo","aragon","perche","ronal"]
print(lista)

lista2 = [2,4,9,4,5]
print(lista2)

lista3 = [True,False,True,True,False]
print(lista3)
```

```python
"""Agrega un elemento al final de la lista."""
lista.append("david")
print(lista)

lista2.append(6)
print(lista2)

lista3.append(True)
print(lista3)
```

```python
"""Elimina todos los elementos de la lista."""
lista.clear()
print(lista)

lista2.clear()
print(lista2)

lista3.clear()
print(lista3)
```

```python
"""Devuelve una copia de la lista"""

lista = ["luque","hugo","aragon","perche","ronal"]
print(lista)

lista.copy()
print(lista)

lista2 = [2,4,9,4,5]
print(lista2)

lista2.copy()
print(lista2)

lista3 = [True,False,True,True,False]
print(lista3)

lista3.copy()
print(lista3)
```

```python
"""Devuelve el número de elementos con el valor especificado"""

print(lista.count("ronal"))

print(lista2.count(4))

print(lista3.count(True))
```

```python
"se le esta añadiendo un elemento iterable al final de la lista no toma numeros ni boleanos"""

lista = ["luque","hugo","aragon","perche","ronal"]
lista.extend(["maicol","bocanegra","esteban"])
print(lista)

lista2 = [2,4,9,4,5]
lista2.extend([3,5,8,7,6,4])
print(lista2)

lista3 = [True,False,True,True,False]
lista3.extend([True,True,False,False])
print(lista3)
```

```python
lista = ["luque","hugo","aragon","perche","ronal"]

lista2 = [2,4,9,4,5]

lista3 = [True,False,True,True,False]

"""Devuelve el índice del primer elemento con el valor especificado"""

print(lista.index("luque"))

print(lista2.index(4))

print(lista3.index(True))
```

```python
"""Añade un elemento en la posición especificada."""
lista.insert(2, "juan")
print(lista)

lista2.insert(4, 10)
print(lista2)

lista3.insert(1, False)
print(lista3)
```

```python
"""Elimina un elemento en la posición especificada."""
lista.pop(2)
print(lista)

lista2.pop(4)
print(lista2)

lista3.pop(1)
print(lista3)
```

```python
"""Elimina el elemento con el valor especificado."""

lista.remove("luque")
print(lista)

lista2.remove(4)
print(lista2)

lista3.remove(True)
print(lista3)
```

```python
"""Invierte el orden de la lista."""

lista.reverse()
print(lista)

lista2.reverse()
print(lista2)

lista3.reverse()
print(lista3)
```

```python
"""Ordena la lista."""

lista.sort()
print(lista)

lista2.sort()
print(lista2)

lista3.sort()
print(lista3)
```

# DICTIONARIO METHODS

```python
"""Elimina todos los elementos del diccionario."""
cliente = {
    "nombre": "maicol",
    "apellido": "garcia",
    "id": 1290219021,
    "registro": True
}

tienda = {
    "vegetal": "zanahoria",
    "frutas": "manzana",
    "id": 1290219021,
    "registro": False
}

zapateria = {
    "zapatos": "nike",
    "camisas": "puma",
    "canitad": 8,
    "registro": True
}

cliente.clear()
print(cliente)

tienda.clear()
print(tienda)

zapateria.clear()
print(zapateria)
```

```python
"""Returns a copy of the dictionary"""
cliente = {
    "nombre": "maicol",
    "apellido": "garcia",
    "id": 1290219021,
    "registro": True
}

tienda = {
    "vegetal": "zanahoria",
    "frutas": "manzana",
    "id": 1290219021,
    "registro": False
}

zapateria = {
    "zapatos": "nike",
    "camisas": "puma",
    "canitad": 8,
    "registro": True
}

copia_cliente = cliente.copy()
print(copia_cliente)

copia_tienda = tienda.copy()
print(copia_tienda)

copia_zapateria = zapateria.copy()
print(copia_zapateria)

```

```python
"""fromkeys() Devuelve un diccionario con las claves y el valor especificados."""
claves = ("nombre", "apellido","id")
valores = ("null")
cliente = dict.fromkeys(claves, valores)
claves = ("vegetal", "frutas","id")
valores = ("null")
tienda = dict.fromkeys(claves, valores)
claves = ("zapatos", "camisas","canitad")
valores = ("null")
zapateria = dict.fromkeys(claves, valores)
print(cliente)
print(tienda)
print(zapateria)
```

```python
"""get() Devuelve el valor de la clave especificada."""
cliente = {
    "nombre": "maicol",
    "apellido": "garcia",
    "id": 1290219021,
    "registro": True
}

tienda = {
    "vegetal": "zanahoria",
    "frutas": "manzana",
    "id": 1290219021,
    "registro": False
}

zapateria = {
    "zapatos": "nike",
    "camisas": "puma",
    "canitad": 8,
    "registro": True
}

print(cliente.get("nombre"))
print(tienda.get("frutas"))
print(zapateria.get("camisas"))
```

```python
"""items() Devuelve una lista que contiene una tupla para cada par clave-valor."""
cliente = {
    "nombre": "maicol",
    "apellido": "garcia",
    "id": 1290219021,
    "registro": True
}

tienda = {
    "vegetal": "zanahoria",
    "frutas": "manzana",
    "id": 1290219021,
    "registro": False
}

zapateria = {
    "zapatos": "nike",
    "camisas": "puma",
    "canitad": 8,
    "registro": True
}

print(cliente.items())
print(tienda.items())
print(zapateria.items())
```

```python
"""keys() Devuelve una lista con las claves del diccionario."""
print(cliente.keys())
print(tienda.keys())
print(zapateria.keys())
```

```python
"""pop() Elimina el elemento con la clave especificada."""
cliente.pop("id")
tienda.pop("registro")
zapateria.pop("canitad")
print(cliente)
print(tienda)
print(zapateria)
```

```python
"""popitem() Elimina el último par clave-valor insertado."""
cliente = {
    "nombre": "maicol",
    "apellido": "garcia",
    "id": 1290219021,
    "registro": True
}

tienda = {
    "vegetal": "zanahoria",
    "frutas": "manzana",
    "id": 1290219021,
    "registro": False
}

zapateria = {
    "zapatos": "nike",
    "camisas": "puma",
    "canitad": 8,
    "registro": True
}
cliente.popitem()
tienda.popitem()
zapateria.popitem()
print(cliente)
print(tienda)
print(zapateria)
```

```python
"""setdefault() Devuelve el valor de la clave especificada. Si la clave no existe: inserta la clave con el valor especificado."""
cliente.setdefault("jugar")
tienda.setdefault("papas")
zapateria.setdefault("reloj")
print(cliente)
print(tienda)
print(zapateria)
```

```python
"""update() Actualiza el diccionario con los pares clave-valor especificados."""
cliente.update({"nombre": "daniel"})
tienda.update({"frutas": "pera"})
zapateria.update({"zapatos": "adidas"})
print(cliente)
print(tienda)
print(zapateria)
```

```python
"""values() Devuelve una lista de todos los valores del diccionario."""
cliente.values()
tienda.values()
zapateria.values()
print(cliente)
print(tienda)
print(zapateria)
```

## TUPLE METHODS

```python
"""Devuelve el número de veces que aparece un valor especificado en una tupla."""
tupla = ("luque","aragon","perche","ronal","luque","hugo")
print(tupla.count("luque"))
camisa = ("nike","puma","adidas","nike","adidas","nike")
print(camisa.count("nike"))
animales = ("gato","perro","gato","gato","gato","gato","gato")
print(animales.count("gato"))
```

```python
"""Busca en la tupla un valor especificado y devuelve la posición donde se encontró."""
print(tupla.index("hugo"))
print(camisa.index("nike"))
print(animales.index("perro"))
```
